```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.loadtxt('ex4Data/ex4x.dat',dtype=np.float32)
y=np.loadtxt('ex4Data/ex4y.dat',dtype=np.float32)
```

```python
plt.scatter(x[:,0],x[:,1],c=y,cmap='coolwarm')

plt.title('data',fontweight='bold')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```

```python
one=np.ones((x.shape[0],3))
```

```python
X=np.concatenate((one,x),axis=1)
```

```python
X[:,1]=x[:,0]
X[:,2]=x[:,1]
X
```

```python
def hypothesis(x,w):
    return 1/(1+np.exp(-np.dot(x,w)))
# Cost Function
def cost(x, y, w):
    return -(1 / x.shape[0]) * np.sum(y*np.log(hypothesis(x,w))+(1-y)*np.log(1-hypothesis(x,w)))
# Gradient Descent
def gradient(x, y, w):
    return (1 / x.shape[0]) * x.T.dot(hypothesis(x, w) - y)
```

```python
w=np.array([0.0, 0.0, 0.0])
costVal=[]
epochs=[]
for i in range(1,10000):
    costVal.append(cost(X,y,w))
    w=w-0.002*gradient(X,y,w)
    epochs.append(i)
```
